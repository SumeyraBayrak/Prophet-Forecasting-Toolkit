<h2 style="font-family: Arial, sans-serif; margin-bottom: 10px;">Python API: Prophet</h2>
<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    <strong>Prophet</strong>, zaman serisi verilerini modellemek ve tahmin etmek için kullanılan bir kütüphanedir ve 
    <strong>scikit-learn (sklearn)</strong> API'sini takip eder. Bu, Prophet'i kullanmanın benzer olduğu anlamına gelir; yani, 
    önce bir Prophet sınıfı oluşturulur, ardından <code>fit</code> ve <code>predict</code> metodları kullanılarak model 
    eğitilir ve tahmin yapılır.
</font>

<h3 style="font-family: Arial, sans-serif; margin-bottom: 10px;">Prophet'e Veri Girişi</h3>
<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Prophet'e veri girişi, iki sütundan oluşan bir <strong>dataframe</strong> şeklinde olmalıdır:
    <ul style="margin-top: 10px; padding-left: 20px;">
        <li>
            <strong>ds</strong>: 
            <ul style="margin-top: 5px; padding-left: 20px;">
                <li>Bu sütun tarih veya zaman damgasını içermelidir.</li>
                <li>Pandas tarafından tanınan bir formatta olmalıdır.</li>
                <li>Genellikle <strong>YYYY-MM-DD</strong> (tarih formatı) veya <strong>YYYY-MM-DD HH:MM:SS</strong> (tarih ve saat formatı) kullanılır.</li>
            </ul>
        </li>
        <li>
            <strong>y</strong>: 
            <ul style="margin-top: 5px; padding-left: 20px;">
                <li>Bu sütun sayısal değerlerden oluşur ve zaman serisinin tahmin edilmek istenen ölçümünü temsil eder.</li>
                <li>Örneğin, günlük satışlar, ziyaretçi sayıları veya herhangi bir ölçüm olabilir.</li>
            </ul>
        </li>
    </ul>
</font>


<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Bir örnek olarak, Peyton Manning'in Wikipedia sayfasının günlük görüntüleme verilerini inceleyelim. Bu veri, R'deki 
    <strong>Wikipediatrend</strong> paketiyle kazındı ve Prophet'in yeteneklerini göstermek için iyi bir örnek sunuyor.
</font>
<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Prophet aşağıdaki özellikleri modelleyebilir:
    <ul style="margin-top: 10px; padding-left: 20px;">
        <li><strong>Mevsimsellik</strong> (örneğin, haftalık ve yıllık döngüler)</li>
        <li><strong>Değişen büyüme oranları</strong></li>
        <li><strong>Özel gün etkileri</strong> (örneğin, Manning'in playoff veya Superbowl maçları)</li>
    </ul>
</font>

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Bu veriyi kullanarak, gelecekteki sayfa görüntülemelerini tahmin edebiliriz. Veri bir CSV dosyası olarak sağlanmıştır. Öncelikle veriyi içe aktaracağız:
</font>

```python
# Python
import pandas as pd
from prophet import Prophet

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df.head()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Modeli eğitmek için önce yeni bir <strong>Prophet</strong> nesnesi oluştururuz. Ayarları yapıcıya (<strong>constructor</strong>) geçiririz. 
    Sonra <code>fit</code> metodunu çağırarak geçmiş veriyi aktarırız. Bu işlem 1-5 saniye sürebilir.
</font>

```python
# Python
m = Prophet()
m.fit(df)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Tahminler, tahmin yapılacak tarihleri içeren <strong>ds</strong> sütununa sahip bir dataframe üzerinde yapılır. 
    Gelecekteki belirli bir süreyi kapsayan bir dataframe oluşturmak için <strong>Prophet.make_future_dataframe</strong> 
    yardımcı metodunu kullanabilirsiniz. Bu metod, varsayılan olarak geçmiş verileri de içerecek şekilde bir dataframe döndürür, 
    böylece modelin eğitildiği verileri ve tahmin edilen geleceği aynı anda görebiliriz.
</font>

```python

# Python
future = m.make_future_dataframe(periods=365)
future.tail()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    <strong>Predict</strong> metodu, gelecekteki her bir tarihe tahmin edilen bir değer atar ve bu değeri <strong>yhat</strong> olarak adlandırır. 
    Eğer geçmiş tarihleri de geçirirseniz, model geçmiş verilerle yaptığı uyumu (<strong>in-sample fit</strong>) gösterir. 
    Bu işlem sonucunda elde edilen <strong>forecast</strong> nesnesi, tahmin sonuçlarını içeren <strong>yhat</strong> sütununun yanı sıra, modelin bileşenleri 
    (örneğin, mevsimsel etkiler) ve belirsizlik aralıkları (güven aralıkları) gibi ek sütunlar da içerir. Bu şekilde, sadece tahmin edilen değeri değil, 
    aynı zamanda modelin çeşitli bileşenlerini ve tahminin güven seviyesini de görebilirsiniz.
</font>

```python
# Python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Tahminleri görselleştirmek için <strong>Prophet.plot</strong> metodunu kullanarak tahmin dataframe'inizi verebilirsiniz.
    Bu metod, modelin tahmin ettiği verileri görsel olarak sunar ve tahmin edilen değerlerin yanı sıra güven aralıklarını da gösterir.
</font>

```python
# Python
fig1 = m.plot(forecast)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Tahmin bileşenlerini görüntülemek için <strong>Prophet.plot_components</strong> metodunu kullanabilirsiniz. 
    Bu yöntem, zaman serisinin trendini, yıllık ve haftalık mevsimselliklerini varsayılan olarak gösterir. 
    Ayrıca, tatiller eklediyseniz, bunlar da görselleştirmede yer alacaktır.
</font>

```python
# Python
fig2 = m.plot_components(forecast)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
    Tahmin ve bileşenlerin etkileşimli bir görselini oluşturmak için <strong>plotly</strong> kullanılabilir. Ancak, plotly'nin 4.0 veya daha yüksek bir sürümünü ayrı olarak yüklemeniz gerekecektir, 
    çünkü bu sürüm Prophet ile birlikte varsayılan olarak yüklenmez. Ayrıca, <strong>notebook</strong> ve <strong>ipywidgets</strong> paketlerini de yüklemeniz gerekir.
</font>

```python
# Python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(m, forecast)
```

```python
# Python
plot_components_plotly(m, forecast)
```
