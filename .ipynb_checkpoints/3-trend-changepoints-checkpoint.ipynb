```python
# Python
import pandas as pd
from prophet import Prophet

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df.head()
```

```python
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
import matplotlib.pyplot as plt

# Prophet modelini oluşturma ve eğitme
m = Prophet(changepoint_prior_scale=0.5)
m.fit(df)

# Tahmin yapma
forecast = m.predict(df)

# Değişim noktalarını al
changepoints = m.changepoints

# Grafik oluşturma
fig = m.plot(forecast)

# Değişim noktalarını çubuklarla ekleme
for cp in changepoints:
    plt.axvline(x=cp, color='grey', linestyle='--', lw=2)  # Değişim noktalarını dikey çizgiyle işaretliyoruz

# Grafik başlıkları
plt.title('Tahmin ve Değişim Noktaları')
plt.xlabel('Tarih')
plt.ylabel('Değer')

# Grafik gösterme
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Örnek DataFrame oluşturma (df'nin içeriğini sağlamak için örnek)
# df = pd.DataFrame({'ds': pd.date_range(start='1/1/2000', periods=1000), 'y': np.random.randn(1000)})

# Prophet modelini oluşturma ve eğitme
m = Prophet()
m.fit(df)

# Gelecek tarihleri oluşturma (365 gün ekleme)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

# Potansiyel değişim noktalarını elde etme
changepoints = m.changepoints  # Prophet'in tespit ettiği değişim noktaları

# Modelin tahmin ettiği trendin değerleri
trend_values = forecast['trend'].values  # Trend değişim değerleri

# Grafik oluşturma
plt.figure(figsize=(10, 6))

# Tarihlere sırasıyla sayılar atama (datetime objelerini sıralı sayılara dönüştürme)
forecast['ds'] = pd.to_datetime(forecast['ds'])
forecast['rank'] = forecast['ds'].rank(method='first').astype(int)

# Trend değişikliklerinin çizimi
plt.plot(forecast['rank'], trend_values, label='Trend', color='green')

# Değişim noktalarını çizme
for changepoint in changepoints:
    changepoint_rank = forecast[forecast['ds'] == changepoint]['rank'].values[0]
    plt.axvline(x=changepoint_rank, color='red', linestyle='--', label='Changepoint')

plt.axhline(0, color='blue', linestyle='--')
plt.xlabel('Date (Rank)')
plt.ylabel('Trend Value')
plt.title('Trend and Potential Changepoints')

# Grafik gösterimi
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Örnek DataFrame oluşturma (df'nin içeriğini sağlamak için örnek)
# df = pd.DataFrame({'ds': pd.date_range(start='1/1/2000', periods=1000), 'y': np.random.randn(1000)})

# Prophet modelini oluşturma ve eğitme
m = Prophet()
m.fit(df)

# Gelecek tarihleri oluşturma (365 gün ekleme)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

# Potansiyel değişim noktalarını elde etme
changepoints = m.changepoints  # Prophet'in tespit ettiği değişim noktaları

# Modelin tahmin ettiği trendin değerleri
trend_values = forecast['trend'].values  # Trend değişim değerleri

# Grafik oluşturma
plt.figure(figsize=(10, 6))

# Tarihlere sırasıyla sayılar atama (datetime objelerini sıralı sayılara dönüştürme)
forecast['ds'] = pd.to_datetime(forecast['ds'])
forecast['rank'] = forecast['ds'].rank(method='first').astype(int)

# Trend değişikliklerinin çubuk grafikle çizimi
plt.bar(forecast['rank'], trend_values, width=1, color='green', alpha=0.6, label='Trend')

# Değişim noktalarını çizme
for changepoint in changepoints:
    changepoint_rank = forecast[forecast['ds'] == changepoint]['rank'].values[0]
    plt.axvline(x=changepoint_rank, color='red', linestyle='--', label='Changepoint')

# Yatay sıfır çizgisi
plt.axhline(0, color='blue', linestyle='--')

# Etiketler ve başlık
plt.xlabel('Date (Rank)')
plt.ylabel('Trend Value')
plt.title('Trend and Potential Changepoints')

# Grafikteki etiketlerin tekrarlanmasını engelleme (aynı etiketlerin birden fazla görünmesini engellemek için)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Grafik gösterimi
plt.show()
```
