```python
# Python
import pandas as pd
from prophet import Prophet

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df.head()
```

```python
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
import matplotlib.pyplot as plt

# Prophet modelini oluturma ve eitme
m = Prophet(changepoint_prior_scale=0.5)
m.fit(df)

# Tahmin yapma
forecast = m.predict(df)

# Deiim noktalarn al
changepoints = m.changepoints

# Grafik oluturma
fig = m.plot(forecast)

# Deiim noktalarn ubuklarla ekleme
for cp in changepoints:
    plt.axvline(x=cp, color='grey', linestyle='--', lw=2)  # Deiim noktalarn dikey izgiyle iaretliyoruz

# Grafik balklar
plt.title('Tahmin ve Deiim Noktalar')
plt.xlabel('Tarih')
plt.ylabel('Deer')

# Grafik gsterme
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# rnek DataFrame oluturma (df'nin ieriini salamak iin rnek)
# df = pd.DataFrame({'ds': pd.date_range(start='1/1/2000', periods=1000), 'y': np.random.randn(1000)})

# Prophet modelini oluturma ve eitme
m = Prophet()
m.fit(df)

# Gelecek tarihleri oluturma (365 gn ekleme)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

# Potansiyel deiim noktalarn elde etme
changepoints = m.changepoints  # Prophet'in tespit ettii deiim noktalar

# Modelin tahmin ettii trendin deerleri
trend_values = forecast['trend'].values  # Trend deiim deerleri

# Grafik oluturma
plt.figure(figsize=(10, 6))

# Tarihlere srasyla saylar atama (datetime objelerini sral saylara dntrme)
forecast['ds'] = pd.to_datetime(forecast['ds'])
forecast['rank'] = forecast['ds'].rank(method='first').astype(int)

# Trend deiikliklerinin izimi
plt.plot(forecast['rank'], trend_values, label='Trend', color='green')

# Deiim noktalarn izme
for changepoint in changepoints:
    changepoint_rank = forecast[forecast['ds'] == changepoint]['rank'].values[0]
    plt.axvline(x=changepoint_rank, color='red', linestyle='--', label='Changepoint')

plt.axhline(0, color='blue', linestyle='--')
plt.xlabel('Date (Rank)')
plt.ylabel('Trend Value')
plt.title('Trend and Potential Changepoints')

# Grafik gsterimi
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# rnek DataFrame oluturma (df'nin ieriini salamak iin rnek)
# df = pd.DataFrame({'ds': pd.date_range(start='1/1/2000', periods=1000), 'y': np.random.randn(1000)})

# Prophet modelini oluturma ve eitme
m = Prophet()
m.fit(df)

# Gelecek tarihleri oluturma (365 gn ekleme)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

# Potansiyel deiim noktalarn elde etme
changepoints = m.changepoints  # Prophet'in tespit ettii deiim noktalar

# Modelin tahmin ettii trendin deerleri
trend_values = forecast['trend'].values  # Trend deiim deerleri

# Grafik oluturma
plt.figure(figsize=(10, 6))

# Tarihlere srasyla saylar atama (datetime objelerini sral saylara dntrme)
forecast['ds'] = pd.to_datetime(forecast['ds'])
forecast['rank'] = forecast['ds'].rank(method='first').astype(int)

# Trend deiikliklerinin ubuk grafikle izimi
plt.bar(forecast['rank'], trend_values, width=1, color='green', alpha=0.6, label='Trend')

# Deiim noktalarn izme
for changepoint in changepoints:
    changepoint_rank = forecast[forecast['ds'] == changepoint]['rank'].values[0]
    plt.axvline(x=changepoint_rank, color='red', linestyle='--', label='Changepoint')

# Yatay sfr izgisi
plt.axhline(0, color='blue', linestyle='--')

# Etiketler ve balk
plt.xlabel('Date (Rank)')
plt.ylabel('Trend Value')
plt.title('Trend and Potential Changepoints')

# Grafikteki etiketlerin tekrarlanmasn engelleme (ayn etiketlerin birden fazla grnmesini engellemek iin)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Grafik gsterimi
plt.show()
```
