```python
import pandas as pd
from prophet import Prophet

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_air_passengers.csv')
```

```python
time_series = pd.Series(df["y"], index=df["ds"])

# Frekansı kontrol et
frequency = pd.infer_freq(time_series.index)
if frequency is None:
    print("Bu veri seti düzensiz bir frekansa sahip.")
else:
    print(f"Frekans: {frequency}")
```

```python
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(50, freq='MS')
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
# Python
m = Prophet(seasonality_mode='multiplicative')
m.fit(df)
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
fig = m.plot_components(forecast)
```

```python
m = Prophet(seasonality_mode='multiplicative')
m.add_seasonality('quarterly', period=91.25, fourier_order=8, mode='additive')
m.add_regressor('regressor', mode='additive')
```
