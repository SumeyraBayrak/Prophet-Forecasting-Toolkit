```python
# Python
import pandas as pd
from prophet import Prophet

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df.head()
```

```python
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
import matplotlib.pyplot as plt

# Prophet modelini oluturma ve eitme
m = Prophet(changepoint_prior_scale=0.5)
m.fit(df)

# Tahmin yapma
forecast = m.predict(df)

# Deiim noktalarn al
changepoints = m.changepoints

# Grafik oluturma
fig = m.plot(forecast)

# Deiim noktalarn ubuklarla ekleme
for cp in changepoints:
    plt.axvline(x=cp, color='grey', linestyle='--', lw=2)  # Deiim noktalarn dikey izgiyle iaretliyoruz

# Grafik balklar
plt.title('Tahmin ve Deiim Noktalar')
plt.xlabel('Tarih')
plt.ylabel('Deer')

# Grafik gsterme
plt.show()
```

```python
from prophet.plot import add_changepoints_to_plot

m = Prophet()
m.fit(df)

future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

fig = m.plot(forecast)

ax = fig.gca()
# Prophet modeli ve tahmin sonular ile figre deiim noktalarn ekleyin
a = add_changepoints_to_plot(fig.gca(), m, forecast)

# Grafii gster
plt.show()
```

```python
m = Prophet(changepoint_prior_scale=0.5)
forecast = m.fit(df).predict(future)
fig = m.plot(forecast)
```
